```python
import pandas as pd
ds = pd.read_csv('general_data.csv')
ds.head()
```

# CHECKING NULL VALUES

```python
ds.isna().sum()
```

# FILLING NULL VALUES WITH MEAN:

```python
a = ds['NumCompaniesWorked'].mean()
ds.fillna(a , inplace=True)
```

```python
ds.isna().sum().sum()
```

# DROPPING UNNECESSARY FEATURES

```python
ds.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'], axis=1 ,inplace=True)
```

```python
ds.columns
```

# CONVERTING TEXT TO NUMBER 

0 = NO, FEMALE, NON TRAVEL

1 = YES, MALE, TRAVEL FREQUENTLY

2 = TRAVEL RARELY



```python
from sklearn.preprocessing import LabelEncoder
ec = LabelEncoder()
ds['Attrition'] = ec.fit_transform(ds['Attrition'])
ds['Gender'] = ec.fit_transform(ds['Gender'])
ds['BusinessTravel'] = ec.fit_transform(ds['BusinessTravel'])
ds['Department'] = ec.fit_transform(ds['Department'])
ds['EducationField'] = ec.fit_transform(ds['EducationField'])
ds['JobRole'] = ec.fit_transform(ds['JobRole'])
ds['MaritalStatus'] = ec.fit_transform(ds['MaritalStatus'])


ds.head()
```

# APPLYING RANDOM FOREST

```python
ds.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=1000 , max_features=2 , oob_score=True)
features = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked',
       'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
rf_model.fit(X=ds[features], y=ds['Attrition'])
```

# ACCURACY OF MODEL:

```python
print(rf_model.oob_score_)
```

# IMPORTANT FEATURES:

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

# THERE ARE 3 IMPORTANT FEATURES:
1] AGE - 0.097

2]MONTHLY INCOME - 0.094

3]TOTAL WORKING YEARS - 0.085

# APPLYING DECISION TREE:

```python
from sklearn import tree
dt_model = tree.DecisionTreeClassifier(max_depth=8 , max_leaf_nodes=10)
predictors = pd.DataFrame([ds['Age'],ds['MonthlyIncome'],ds['TotalWorkingYears']]).T

dt_model.fit(X=predictors , y=ds['Attrition'])

with open('attrition.dot','w') as f:
    f=tree.export_graphviz(dt_model,feature_names = ['Age','MonthlyIncome','TotalWorkingYears'],out_file=f)
```

# CHECKING ACCURACY OF MODEL:

```python
dt_model.score(X=predictors , y=ds['Attrition'])
```
